```python
%matplotlib inline

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = '../DATA/00-puppy.jpg'
img = cv2.imread(file_path)
img.shape
```

```python
plt.imshow(img)
```

```python
img2 = img.copy()
img2[:,:,1:] = 0
plt.imshow(img2)
```

```python
img3 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img3)
```

```python
img_gray = cv2.imread(file_path,cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray, cmap='gray')
```

```python
img_resize = cv2.resize(img, [500,500])
plt.imshow(img_resize)
```

```python
h_ratio = 0.5
w_ratio = 0.5
img_rdc = cv2.resize(img3,(0,0),img3,w_ratio,h_ratio)
plt.imshow(img_rdc)
```

```python
img_flip = cv2.flip(img3, 0)
plt.imshow(img_flip)
```

```python
fig = plt.figure(figsize=(10,4))
ax = fig.add_subplot(111)
ax.imshow(img3)
```

```python
img_sml = cv2.resize(img,(0,0),img3,w_ratio,h_ratio)
while True:
    cv2.imshow('Perro', img_sml)

    key = cv2.waitKey(1)
    if key == ord('q'):
        break
cv2.destroyAllWindows()
```
