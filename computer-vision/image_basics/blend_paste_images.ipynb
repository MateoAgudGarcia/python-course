```python
%matplotlib inline

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('../DATA/dog_backpack.png')
img2 = cv2.imread('../DATA/watermark_no_copy.png')
# BGR to RGB
dog_rgb = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
water_rgb = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
#RESIZE 
wmk_rsz = cv2.resize(water_rgb, (dog_rgb.shape)[:2][::-1])
```

```python
wmk_gray = cv2.cvtColor(wmk_rsz, cv2.COLOR_RGB2GRAY)
plt.imshow(wmk_gray, cmap='gray')
```

```python
mask_watermark = cv2.bitwise_not(wmk_gray)
plt.imshow(mask_watermark, cmap='gray')
```

```python
white_background = np.full(dog_rgb.shape,255)
white_background.shape
```

```python
background = cv2.bitwise_or(white_background,white_background,mask=mask_watermark)
plt.imshow(background)
```

```python
watermark = cv2.bitwise_or(wmk_rsz,wmk_rsz,mask=mask_watermark)
plt.imshow(watermark)
```

```python
picture_not_copy = cv2.bitwise_or(watermark,dog_rgb)
plt.imshow(picture_not_copy)
```
