```python
%matplotlib inline

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../DATA/rainbow.jpg')
rainbow = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rainbow)
```

```python
img_gray = cv2.imread('../DATA/rainbow.jpg',0)
plt.imshow(img_gray, cmap='gray')
```

```python
ret, threshold_binary = cv2.threshold(img_gray,180,255,cv2.THRESH_BINARY)
plt.imshow(threshold_binary,cmap='gray') 
```

```python
ret, threshold_tozero = cv2.threshold(img_gray,120,255,cv2.THRESH_TOZERO)
plt.imshow(threshold_tozero,cmap='gray')
```

```python
crossword_gray = cv2.imread('../DATA/crossword.jpg',0)
plt.imshow(crossword_gray, cmap='gray')
```

```python
crossword_gray.max()
```

```python
ret, threshold_binary = cv2.threshold(crossword_gray,200,255,cv2.THRESH_BINARY)
plt.imshow(threshold_binary,cmap='gray')
```

```python
threshold_adaptative = cv2.adaptiveThreshold(crossword_gray,200,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,5,6)
plt.imshow(threshold_adaptative,cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
show_pic(threshold_adaptative)
```
