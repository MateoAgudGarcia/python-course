```python
%matplotlib inline

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('../DATA/dog_backpack.png')
img2 = cv2.imread('../DATA/watermark_no_copy.png')
```

```python
dog_rgb = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
water_rgb = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(dog_rgb)
```

```python
plt.imshow(water_rgb)
```

```python
wmk_rsz = cv2.resize(water_rgb, (dog_rgb.shape)[:2][::-1])
blended = cv2.addWeighted(dog_rgb,0.5,wmk_rsz,0.5,0)
plt.imshow(blended)
```

```python
wmk_rsz[wmk_rsz == 255 ] = 0
plt.imshow(wmk_rsz)
```
