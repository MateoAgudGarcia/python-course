```python
import cv2
```

```python
cap = cv2.VideoCapture(0)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# TOP LEFT CORNER

x = width // 2
y = height // 2

# Height and Width for RECTANGLE

w = width // 4
h = height // 4

# BOTTOM RIGHT CORNER (x+w , y+h)

while(1):
    ret, frame = cap.read()

    cv2.rectangle(frame, (x,y), (x+w,y+h), (0,0,255),10)
    cv2.imshow('Camara',frame)

    if cv2.waitKey(20) & 0xFF == 27:  
        break  

cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

def draw_rectangle(event,x,y,flags,param):
    global pt1, pt2, topLeft_clicked, bottonRight_clicked

    if event == cv2.EVENT_LBUTTONDOWN:
        if topLeft_clicked == True and bottonRight_clicked == True:
            pt1 = (0,0)
            pt2 = (0,0)
            topLeft_clicked = False
            bottonRight_clicked = False

        if topLeft_clicked == False:
            pt1 = (x,y)
            topLeft_clicked = True

        elif bottonRight_clicked == False:
            pt2 = (x,y)
            bottonRight_clicked = True

pt1 = (0,0)
pt2 = (0,0)
topLeft_clicked = False
bottonRight_clicked = False

camera_window = 'Event Camera'
cv2.namedWindow(camera_window)
cv2.setMouseCallback(camera_window, draw_rectangle)


while(1):  
    ret,frame = cap.read()

    if topLeft_clicked:
        cv2.circle(frame, pt1,3,(255,0,0),-1)
    if topLeft_clicked and bottonRight_clicked:
        cv2.rectangle(frame,pt1,pt2,(255,0,0),5)

    cv2.imshow(camera_window, frame)
    
    if cv2.waitKey(20) & 0xFF == 27:  
        break  

cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

def draw_rectangle(event,x,y,flags,param):
    global center, lenght, topLeft_clicked, bottonRight_clicked

    if event == cv2.EVENT_LBUTTONDOWN:
        if topLeft_clicked == True and bottonRight_clicked == True:
            center = (0,0)
            radius = (0,0)
            lenght = 0
            topLeft_clicked = False
            bottonRight_clicked = False

        if topLeft_clicked == False:
            center = (x,y)
            topLeft_clicked = True

        elif bottonRight_clicked == False:
            radius = (x,y)
            lenght = int((((radius[1] - center[1] )**2) + ((radius[0] - center[0] )**2) )**0.5)
            bottonRight_clicked = True

center = (0,0)
radius = (0,0)
lenght = 0
topLeft_clicked = False
bottonRight_clicked = False

camera_window = 'Event Camera'
cv2.namedWindow(camera_window)
cv2.setMouseCallback(camera_window, draw_rectangle)


while(1):  
    ret,src = cap.read()

    frame = cv2.flip(src,1)

    if topLeft_clicked and bottonRight_clicked:
        cv2.circle(frame,center,lenght,(0,0,255),5)

    cv2.imshow(camera_window, frame)
    
    if cv2.waitKey(20) & 0xFF == 27:  
        break  

cap.release()
cv2.destroyAllWindows()
    
```
